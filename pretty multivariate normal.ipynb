```r
# libraries
library(dplyr)
library(ggplot2)
library(ggthemes)
library(mnormt)
```

```r
# generate data
mu <- c(2,2)
sigma <- matrix(c(2,2,2.8,5), ncol=2, byrow =T)
x <- rmnorm(1000, mu, varcov=sigma) %>% data.frame()
head(x)
```

```r
# theme
theme_set(theme_tufte() + theme(legend.position = 'n'))

# scatter plot
g <- ggplot(data = x, aes(X1, X2, colour=X2))+ 
    geom_point(alpha=.6) + 
    geom_density_2d(alpha=.4)
g 
```

```r
# densities
g2 <- ggplot(data = x, aes(X1, fill = 1)) + 
    geom_density(alpha = .2)

g3 <- ggplot(data = x, aes(X2, fill = 1)) + 
    geom_density(alpha = .2) + 
    rotate()
```

```r
# multiple plots
library(ggpubr)
ggarrange(g2, NULL, g, g3, 
          ncol = 2, nrow = 2,  
          align = "hv", widths = c(2, 1), 
          heights = c(1, 2))
```
